```python
import pandas as pd

# Sample DataFrame
visa_df = pd.read_csv(r'C:\Users\ankus\Documents\Naresh IT Ankush\Datasets\Visadataset - Visadataset.csv')


# Step 1: Convert series types to dictionary
dtype_dict = visa_df.dtypes.to_dict()

# Initialize lists for categorical and numerical columns
categorical_columns = []
numerical_columns = []

# Step 2: Iterate through the dictionary and apply conditions
for column, dtype in dtype_dict.items():
    if pd.api.types.is_numeric_dtype(dtype):
        numerical_columns.append(column)
    else:
        categorical_columns.append(column)

# Output the results
print("Categorical Columns:", categorical_columns)
print("Numerical Columns:", numerical_columns)
```

```python
import pandas as pd

# Sample DataFrame
visa_df = pd.read_csv(r'C:\Users\ankus\Documents\Naresh IT Ankush\Datasets\Visadataset - Visadataset.csv')

# Step 1: Convert series types to dictionary
dtype_dict = visa_df.dtypes.to_dict()

# Initialize lists for categorical and numerical columns
categorical_columns = []
numerical_columns = []

# Step 2: Iterate through the dictionary and apply conditions
for column, dtype in dtype_dict.items():
    if pd.api.types.is_numeric_dtype(dtype):
        numerical_columns.append(column)
    else:
        categorical_columns.append(column)

# Create separate DataFrames
categorical_df = visa_df[categorical_columns]
numerical_df = visa_df[numerical_columns]

# Output the results
print("Categorical Columns:", categorical_columns)
print("Numerical Columns:", numerical_columns)

# Display DataFrames

categorical_df,numerical_df
```
